# BIG FOOT DATA SECTION
-imported both unclean csv files and merged them into one dataframe, cleaned dataframe and exported as csv and pkl

```python
bf_loc = pd.read_csv("..//Data Cleaning/resources/uncleaned data/timothyrenner-bfro-sightings-data/bfro_locations.csv")
bfLocations = pd.DataFrame(bf_loc)
bfLocations.head()
```

```python
bfLocations.to_pickle("..//Data Cleaning/resources/uncleaned data/timothyrenner-bfro-sightings-data/uncleaned_BFLocations.pkl")
```

```python
bf_geo = pd.read_csv("..//Data Cleaning/resources/uncleaned data/timothyrenner-bfro-sightings-data/bfro_reports_geocoded.csv")
bfGeocode = pd.DataFrame(bf_geo)
bfGeocode.head()
```

```python
bfGeocode.to_pickle("..//Data Cleaning/resources/uncleaned data/timothyrenner-bfro-sightings-data/uncleaned_BFGeocode.pkl")
```

```python
# bfGeocode.columns
bfGeocode["number"] = bfGeocode["number"].astype(int)
bfGeocode.info()
```

```python
# bfLocations.info()
bfData = bfLocations.merge(bfGeocode, how='inner')
bfData.head()
```

```python
bfData = bfData.drop(columns=['uv_index', 'precip_intensity', 'precip_type', 'pressure', 'precip_probability', 'temperature_low', 'temperature_high', 'temperature_mid', 
                            'dew_point', 'humidity', 'cloud_cover', 'conditions', 'uv_index', 'visibility', 'wind_bearing','wind_speed', 'county'])
bfData.head()
```

```python
bfData = bfData.drop(columns=['season', 'moon_phase', 'geohash'])
bfData.head()
```

```python
bfData['timestamp'] = pd.to_datetime(bfData['timestamp'])
bfData.info()
```

```python
bfData = bfData.drop(columns=['date','classification'])
bfData.head()
```

```python
bfData = bfData.dropna()
bfData.head()
```

```python
# remove all coordinates outside of the united states, alaska, and hawaii
USA_BF_CONT = bfData.loc[((bfData['latitude'] >= 24.0) & (bfData['latitude'] <= 49.0) & 
                            (bfData['longitude'] <= -66.0) & (bfData['longitude'] >= -125.0)) | 
                            ((bfData['latitude'] >= 55.0) & (bfData['latitude'] <= 72.0) & 
                            (bfData['longitude'] <= -130) & (bfData['longitude'] >= -180.0)) |
                            ((bfData['latitude'] >= 19.0) & (bfData['latitude'] <= 29.0) & 
                            (bfData['longitude'] <= -155.0) & (bfData['longitude'] >= -179.0))]
USA_BF_CONT.head()
```

```python
USA_BF_CONT.to_pickle("..//Data Cleaning/resources/bigfoot_data.pkl")
```

```python
USA_BF_CONT.to_csv("..//Data Cleaning/resources/bigfoot_data.csv")
```

## UFO DATA SECTION
-imported the 3 seperated uncleaned csv files and exported them all as pkl, merged the upper and middle years so we have a data set >= the year 2000, cleaned data and exported as csv and pkl

```python
ufo_bottom = pd.read_csv("..//Data Cleaning/resources/uncleaned data/timothyrenner-ufo-sightings/ufoLower.csv")
ufo_Lower = pd.DataFrame(ufo_bottom)
ufo_Lower.head()
```

```python
ufo_Lower.to_pickle("..//Data Cleaning/resources/uncleaned data/timothyrenner-ufo-sightings/uncleanedUFO_LowerYears.pkl")
```

```python
ufo_mid = pd.read_csv("..//Data Cleaning/resources/uncleaned data/timothyrenner-ufo-sightings/ufoMiddle.csv")
ufo_Middle = pd.DataFrame(ufo_mid)
ufo_Middle.head()
```

```python
ufo_Middle.to_pickle("..//Data Cleaning/resources/uncleaned data/timothyrenner-ufo-sightings/uncleanedUFO_MiddleYears.pkl")
```

```python
ufo_top = pd.read_csv("..//Data Cleaning/resources/uncleaned data/timothyrenner-ufo-sightings/ufoUpper.csv")
ufo_Upper = pd.DataFrame(ufo_top)
ufo_Upper.head()
```

```python
ufo_Upper.to_pickle("..//Data Cleaning/resources/uncleaned data/timothyrenner-ufo-sightings/uncleanedUFO_UpperYears.pkl")
```

```python
ufoReports = pd.concat([ufo_Middle, ufo_Upper])
ufoReports.head()
```

```python
ufoReports = ufoReports.drop(columns=['report_link', 'posted', 'city', 'stats', 'summary'])
ufoReports.head()
```

```python
ufoReports["date_time"] = pd.to_datetime(ufoReports["date_time"])
ufoReports.info()
```

```python
ufoReports = ufoReports.rename(columns={'date_time': 'datetime', 'city_latitude': 'latitude', 'city_longitude': 'longitude', 'text': 'description'})

ufoReports.head()
```

```python
ufo_clean = ufoReports.dropna()
ufo_clean.head()
```

```python
ufo_clean.info()
```

```python
ufo_sorted = ufo_clean.sort_values('datetime')
ufo_sorted = ufo_sorted.reset_index(drop=True)
ufo_sorted.head()
```

```python
# ufo_sorted.columns

ufo_sorted = ufo_sorted.drop(columns=['Unnamed: 0'])
ufo_sorted.head()
```

```python
# remove all coordinates outside of the united states, alaska, and hawaii

UFO_USA_CONT = ufo_sorted.loc[((ufo_sorted['latitude'] >= 24.0) & (ufo_sorted['latitude'] <= 49.0) & 
                            (ufo_sorted['longitude'] <= -66.0) & (ufo_sorted['longitude'] >= -125.0)) | 
                            ((ufo_sorted['latitude'] >= 55.0) & (ufo_sorted['latitude'] <= 72.0) & 
                            (ufo_sorted['longitude'] <= -130) & (ufo_sorted['longitude'] >= -180.0)) |
                            ((ufo_sorted['latitude'] >= 19.0) & (ufo_sorted['latitude'] <= 29.0) & 
                            (ufo_sorted['longitude'] <= -155.0) & (ufo_sorted['longitude'] >= -179.0))]
UFO_USA_CONT.head()
```

```python
UFO_USA_CONT.to_pickle("..//Data Cleaning/resources/ufo_data.pkl")
```
